# Data Prep

## Reshaping Data

```python
import pandas as pd
```

```python
songs_dict = {
                'Customer': ['Aria', 'Aria', 'Aria', 'Chord', 'Chord', 'Harmony', 'Harmony', 'Harmony', 'Melody', 'Reed'],
                'Genre': ['Pop', 'Indie', 'Rock', 'Pop', 'Indie', 'Pop', 'Indie', 'Rock', 'Rock', 'Rock'],
                '# Songs': [50, 48, 1, 15, 36, 10, 5, 3, 2, 5]
             }

df = pd.DataFrame(songs_dict)
df
```

```python
# groub by
(df.groupby(["Customer"])
   .agg(
       TotalNumSong =('# Songs', 'sum')
       )
   .reset_index())
```

```python
customers_songs = (df.groupby(["Customer"])
                     .agg(TotalNumSong =('# Songs', 'sum'))
                     .reset_index()
                  )
```

```python
customers_songs
```

```python
# pivot
(df.pivot(index=['Customer'], columns=['Genre'], values='# Songs')
   .fillna(0)
   .reset_index()
)
```

```python
customers_genres = (
    df.pivot(index=['Customer'], columns=['Genre'], values='# Songs')
   .fillna(0)
   .reset_index()
)
```

```python
customers_genres
```
